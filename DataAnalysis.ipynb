```python
import pandas as pd
from scipy.stats import trim_mean
```

```python
data = {
    'State': ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware'],
    'Population': [4779736, 710231, 6392017, 2915918, 37253956, 5029196, 3574097, 897934],
    'Murder_Rate': [5.7, 5.6, 4.7, 5.6, 4.4, 2.8, 2.4, 5.8]
}
```

```python
state = pd.DataFrame(data)
```

```python
print("--- Data Table ---")
print(state.head(8))
print("\n")
```

```python
# the mean

pop_mean = state["Population"].mean()
print(f"Mean Population: {pop_mean:,.0f}")
```

```python
# the median

pop_median = state["Population"].median()
print(f"Median Population: {pop_median:,.0f}")
```

```python
# trimmed mean

pop_trim = trim_mean(state["Population"] , 0.1)
print(f"Trimmed Mean Population: {pop_trim:,.0f}")
```

```python
sorted_population = state["Population"].sort_values().reset_index(drop=True)
print(f"Sorted Population data:\n{sorted_population}")
```

As you can see from the sorted data, since 0.8 data points need to be trimmed from each side, and no whole data points are trimmed, the mean of the entire dataset is returned.
