```python
import numpy as np
import matplotlib.pyplot as plt


from skimage import data , color , filters
```

Import all the needed packages


```python
def show_image(image , title='Image', cmap_type='gray'):
    plt.imshow(image , cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

A fuction that helps us to plot our images

```python
# load a sample image

image = data.rocket()
show_image(image , "Rocket Image", cmap_type=None)
```

rgb channels


```python
red = image[ : , : , 0]
green = image[ : , : , 1]
blue = image[ : , : , 2]
```

Visualising each channel

```python
show_image(red , "Red Channel")
show_image(green , "Green Channel")
show_image(blue , "Blue Channel")
```

RGB to grayscale

```python
grayscale = color.rgb2gray(image)
show_image(grayscale , "Grayscale")
```

```python
# vertically flipped
vertical_flip = np.flipud(image)
show_image(vertical_flip , "Vertical Flip")
```

```python
# Horizontal flip
horizontal_flip = np.fliplr(image)
show_image(horizontal_flip , "Horizontal Flip")
```

```python
# Grayscale histogram
plt.hist(grayscale.ravel(), bins=256)
plt.title("Grayscale Histogram")
plt.show()
```

RGB Histogram

```python
# Red channel histogram

plt.hist(red.ravel() , bins=256)
plt.title("Red Channel Histogram")
plt.show()
```

```python
# Green channel histogram


plt.hist(green.ravel() , bins=256)
plt.title("Green Channel Histogram")
plt.show()
```

```python
# Blue channel histogram


plt.hist(blue.ravel() , bins=256)
plt.title("Blue Channel Histogram")
plt.show()
```
